```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def make_filter(r):
    Y, X = np.mgrid[:2*r+1, :2*r+1]
    dist = (Y - r)**2 + (X - r)**2
    mask = dist <= r**2
    mask = mask/np.sum(mask)
```

```python
make_filter(100)
```

```python
def min_dist(x, y, d):
    h, w = 100, 80
    img = np.zeros((h,w))
    img[x, y] = 1
    points = set()
    for i in range(d+1):
        j = np.round((d**2-i**2)**(0.5))
        points.add((j, i))
        points.add((i, j))
    points = np.array(list(points)).astype(np.int)
    p1 = points + [x, y]
    p2 = points * [-1, 1] + [x, y]
    p3 = points * [1, -1] + [x, y]
    p4 = points * [-1, -1] + [x, y]
    points = np.concatenate([p1,p2,p3,p4])
    final_points = []
    for p in points:
#         if p[0] >= 0 and p[0] < 100:
#             if p[1] >= 0 and p[1] < 80:
        img[p[0], p[1]] = 1
    plt.imshow(img)
    plt.show()
```

```python
min_dist(25, 35, 45)
```
